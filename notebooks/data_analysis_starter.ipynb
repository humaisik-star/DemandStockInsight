# Starter Data Analysis Notebook

Use this notebook to load your dataset, explore its structure, and run initial analysis. Place your data file in the `data/` folder and update the filename below.

```python
import pandas as pd
from pathlib import Path

data_path = Path('..') / 'data' / 'your_dataset.csv'
print('Data file:', data_path)

# Replace 'your_dataset.csv' with the actual file name
df = pd.read_csv(data_path)
df.head()
```

## Initial dataset summary

Run the next cell to see the dataset shape and basic info.

```python
print('Shape:', df.shape)
print('
Columns:')
print(df.columns.tolist())
print('
Info:')
print(df.info())
print('
Describe:')
print(df.describe(include='all'))
```

## Basic visual checks

Run this cell if your dataset contains numeric columns.

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
print('Numeric columns:', numeric_cols)

if len(numeric_cols) > 0:
    plt.figure(figsize=(12, 6))
    sns.heatmap(df[numeric_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm')
    plt.title('Numeric feature correlation matrix')
    plt.show()

    df[numeric_cols].hist(figsize=(12, 8), bins=20)
    plt.tight_layout()
    plt.show()
else:
    print('No numeric columns found for visualization.')
```
