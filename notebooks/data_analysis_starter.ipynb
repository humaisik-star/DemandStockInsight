# Starter Data Analysis Notebook

Use this notebook to load your dataset, explore its structure, and run initial analysis. Place your data file in the `data/` folder and update the filename below.

```python
import pandas as pd
from pathlib import Path

base_dir = Path.cwd()
data_dir = base_dir / 'data'
if not data_dir.exists():
    data_dir = base_dir.parent / 'data'

candidates = [
    data_dir / 'inventory_demand_forecasting_dataset.csv',
    data_dir / 'demand_forecasting.csv',
]

data_path = next((p for p in candidates if p.exists()), None)
if data_path is None:
    raise FileNotFoundError(f'No dataset file found in {data_dir}')

print('Data file:', data_path)
df = pd.read_csv(data_path)
df.head()
```

## Initial dataset summary

Run the next cell to see the dataset shape and basic info.

```python
print('Shape:', df.shape)
print('Columns:')
print(df.columns.tolist())
print('Info:')
print(df.info())
print('Describe:')
print(df.describe(include='all'))
```

## Basic visual checks

Run this cell if your dataset contains numeric columns.

```python
import matplotlib.pyplot as plt

try:
    import seaborn as sns
except ModuleNotFoundError:
    sns = None

numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
print('Numeric columns:', numeric_cols)

if len(numeric_cols) > 0:
    corr = df[numeric_cols].corr()

    plt.figure(figsize=(12, 6))
    if sns is not None:
        sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
    else:
        plt.imshow(corr, cmap='coolwarm')
        plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha='right')
        plt.yticks(range(len(corr.index)), corr.index)
        for i in range(corr.shape[0]):
            for j in range(corr.shape[1]):
                plt.text(j, i, f'{corr.iloc[i, j]:.2f}', ha='center', va='center')
    plt.title('Numeric feature correlation matrix')
    plt.show()

    plt.figure(figsize=(12, 8))
    df[numeric_cols].hist(bins=20)
    plt.tight_layout()
    plt.show()
else:
    print('No numeric columns found for visualization.')
```
